<h2> Analysis using small vox datatset

Read in the data & store in a list of dictionaries (calling it vox_data)

```python
import csv

def prepare_datasets(file_path):
    with open(file_path) as infile:
        reader = csv.DictReader(infile, delimiter=',')
        list_of_dicts = [dict(r) for r in reader]
        
    return list_of_dicts

vox_data = prepare_datasets("vox-disinfo.csv")
```

```python
from pprint import pprint

date_counts = {}
for i in vox_data:
    #print(i)
    date = i['Date']
    #print(date)
    if date in date_counts: date_counts[date] += 1
    else: date_counts[date] = 1

pprint(date_counts)
```
